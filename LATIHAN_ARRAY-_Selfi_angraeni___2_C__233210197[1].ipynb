```python
# perkalian matriks array

import numpy as np

a = np.array([[[12,3], [4,5],[6,7]]])
b = np.array([[[8,9], [1,2], [3,4]]])

kali = a*b
print(kali)
```

```python
# pengurangan matriks array

import numpy as np

a = np.array([[[5,7], [4,3],[9,11]]])
b = np.array([[[2,8], [5,6], [8,5]]])

kurang = a-b
print(kurang)
```

```python
# transpose matriks array

import numpy as np

a = np.array([[6,8,7], [9,3,4], [1,3,2]])
b = np.array([[6,7,7], [3,5,3], [8,6,4]])

transpose_a = np.transpose (a)
transpose_b = np.transpose (b)
print (transpose_a)
print ( transpose_b)
```

```python
# inverse matriks array

import numpy as np

a = np.array([[3,1], [3,3]])
b = np.array([[8,9], [ 3,2]])

inverse_a = np.linalg.inv(a)
inverse_b = np.linalg.inv(b)
print("\nInverse matrkis a:")
print(inverse_a)
print("\nInverse matrkis b:")
print(inverse_b)

```

```python
# determinan matriks array

import numpy as np

a = np.array([[3,9], [9,5]])
b = np.array([[5,6], [ 4,1]])

determinan_a = np.linalg.det(a)
determinan_b = np.linalg.det(b)
print("Determinan dari matriks a:", determinan_a) 
print("Determinan dari matriks b:", determinan_b)
```
